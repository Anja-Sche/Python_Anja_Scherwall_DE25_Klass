# Matplotlib
- in this course, focus on the distinction between two ways of writing matplotlib code
- in this course, focus on the components of a matplotlib figure

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# generate data x
x = np.linspace(-5, 5, 100)
x
```

```python
# generate data f
f = np.sin(x)
f
```

##### This first approach - functional/matlab style

```python
plt.figure()
plt.plot(x,f)

plt.title("sin (X)")
plt.xlabel("x")
plt.ylabel("f")
```

##### The second approach - OOP

```python
fig = plt.figure() # create a figure
ax = plt.axes() # on the figure, create an axe
ax.plot(x,f) # on the axe, create a plot
ax.set(xlim=(-10,10), ylim=(-5,5), xlabel="x", ylabel="y") # continue to modify the axe objectin different aspects

# check out more about the compinents of matplotlib figure (Sök matplotlib disection, klicka på anatomy of a figure)
```
